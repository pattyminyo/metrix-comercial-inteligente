```python
print("📊 Metrix Comercial Inteligente iniciado!")
print("Digite sua pergunta comercial.")
print("Para encerrar, digite: sair\n")

while True:
    pergunta = input("Pergunta: ")
    pergunta_lower = pergunta.lower().strip()

    if pergunta_lower == "sair":
        print("\nEncerrando o Metrix Comercial Inteligente. Até a próxima!")
        break

    if "aumentar vendas" in pergunta_lower or "melhorar vendas" in pergunta_lower:
        resposta = "Para aumentar vendas, acompanhe conversão, ticket médio, produtividade da equipe e desempenho por canal."
    elif "cancelamento" in pergunta_lower or "cancelar" in pergunta_lower:
        resposta = "Para reduzir cancelamentos, fortaleça o pós-venda, monitore os principais motivos de saída e atue nas causas de insatisfação dos clientes."
    elif "churn" in pergunta_lower:
        resposta = "Churn é a taxa de perda de clientes em um período. Quanto menor esse indicador, melhor tende a ser a retenção."
    elif "cac" in pergunta_lower:
        resposta = "CAC é o custo de aquisição de clientes. Ele mostra quanto a empresa investe para conquistar cada novo cliente."
    elif "ticket médio" in pergunta_lower or "ticket medio" in pergunta_lower:
        resposta = "Ticket médio é o valor médio vendido por cliente ou pedido em um período."
    elif "convers" in pergunta_lower or "fechar venda" in pergunta_lower:
        resposta = "A taxa de conversão mede a proporção entre oportunidades geradas e vendas efetivadas."
    elif "meta" in pergunta_lower or "bater meta" in pergunta_lower or "resultado" in pergunta_lower:
        resposta = "Para bater metas, acompanhe volume vendido, taxa de conversão, ticket médio e evolução diária ou semanal."
    elif "kpi" in pergunta_lower or "indicador" in pergunta_lower:
        resposta = "Os principais KPIs comerciais incluem conversão, churn, CAC, ticket médio, receita e produtividade."
    elif "produtividade" in pergunta_lower:
        resposta = "A produtividade comercial pode ser avaliada por volume vendido, contatos realizados, conversão e resultado por vendedor."
    elif "receita" in pergunta_lower:
        resposta = "Receita representa o valor total gerado pelas vendas em um período. Ela deve ser analisada junto com ticket médio e volume vendido."
    else:
        resposta = "Posso ajudar com temas como vendas, churn, cancelamento, CAC, conversão, metas, produtividade, receita e ticket médio."

    print("\n💡 Resposta Inteligente:")
    print(resposta)
    print("\n" + "-" * 60 + "\n")
```
